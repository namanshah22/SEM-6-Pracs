```python
import math

class ColumnCipher:
    def rank(self, key):
        sb = []
        ch = sorted(key.lower())
        for i in range(len(key)):
            for j in range(len(key)):
                if key.lower()[i] == ch[j]:
                    ch[j] = '@'
                    sb.append(str(j + 1))
                    break
        return ''.join(sb)

    def encrypt_matrix(self, text, n):
        array_list = []
        row = math.ceil(len(text) / n)
        for i in range(row):
            array = []
            for j in range(n):
                if i * n + j < len(text):
                    array.append(text[i * n + j])
            array_list.append(array)
        while len(array_list[-1]) < n:
            array_list[-1].append('@')
        return array_list

    def encrypt(self, rank, text):
        map_ = {}
        array_list = self.encrypt_matrix(text.lower(), len(rank))
        sb = []
        for i in range(len(rank)):
            map_[int(rank[i])] = i
        for i in range(len(rank)):
            for j in range(math.ceil(len(text) / len(rank))):
                sb.append(array_list[j][map_[i + 1]])
        return ''.join(sb)

    def decrypt_matrix(self, cipher, map_, n):
        row = math.ceil(len(cipher) / n)
        ch = [['' for _ in range(n)] for _ in range(row)]
        for i in range(n):
            for j in range(row):
                ch[j][map_[i + 1]] = cipher[i * row + j]
        return ch

    def decrypt(self, rank, cipher):
        map_ = {}
        sb = []
        for i in range(len(rank)):
            map_[int(rank[i])] = i
        ch = self.decrypt_matrix(cipher.lower(), map_, len(rank))
        for chars in ch:
            for c in chars:
                sb.append(c)
        return ''.join(sb).replace("@", "")


def main():
    column_cipher = ColumnCipher()
    print("Enter the key: ")
    key = input().strip()
    print("Enter the text to be ciphered: ")
    text = input().strip()
    rank = column_cipher.rank(key.lower())
    encrypted_text = column_cipher.encrypt(rank, text)
    decrypted_text = column_cipher.decrypt(rank, encrypted_text)
    print("The rank of the key", key, "is:", rank)
    print("The encrypted text is:", encrypted_text)
    print("The decrypted text is:", decrypted_text)


if __name__ == "__main__":
    main()
```
